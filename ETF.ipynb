```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

priceETF = 1450
priceETF_I = 1700
money = 1

x = np.linspace(1100,2000)

def profitETF(currentPrice, etfPrice):
    etfPrice = float(etfPrice)
    return (currentPrice - etfPrice) / etfPrice

def profitInverse(currentPrice, inversePrice):
    inversePrice = float(inversePrice)
    return (inversePrice - currentPrice) / inversePrice

profit_ETF = profitETF(x, priceETF) * money
profit_I = profitInverse(x, priceETF_I) * money

fig = plt.figure(figsize=[11,6])
ax = plt.axes()

ax.plot(x, profit_ETF)
ax.plot(x, profit_I)
ax.plot(x, profit_ETF + profit_I)

ax.legend(('ETF','Inverse', 'Both'))
ax.set_title("E: " + str(priceETF) + " ~ I: " + str(priceETF_I))
ax.grid(True)
fig.patch.set_facecolor((0.9, 0.9, 0.9))
```
